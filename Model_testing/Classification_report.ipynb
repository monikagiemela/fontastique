```python
import os
import pandas as pd
import numpy as np
from tensorflow import keras
import tensorflow as tf

from sklearn.metrics import classification_report
```

```python
clf = keras.models.load_model(os.path.join("app", "static", "top_model.h5"))
```

```python
test = tf.keras.utils.image_dataset_from_directory(os.path.join("/content", "drive", "MyDrive", "AI", "fonts", "test"),
                                                   shuffle=True,
                                                   image_size=(105,105),
                                                   batch_size=32,
                                                   labels='inferred',
                                                   label_mode='int')
```

```python
autotune = tf.data.AUTOTUNE
test = test.prefetch(buffer_size=autotune)
```

```python
loss, accuracy = clf.evaluate(test)
```

```python
with open(os.path.join("/content", "drive", "MyDrive", "AI", "fonts", "models_results", "model_testing_results.txt"), "a") as file:
    file.write(
f"""Model_3_3+ {loss, accuracy}\n""")
```

```python
predictions = clf.predict(test, batch_size=32)
```

```python
# assigning true and predicted value for each image
y_pred = []
y_true = []

for image, lable in test:
  y_true.append(lable)
  preds = clf.predict(image)
  y_pred.append(np.argmax(preds, axis=-1))
  with open(os.path.join("/content", "drive", "MyDrive", "AI", "fonts", "models_results", "model_3_3", "model_3_3_biger_dataset_predictions.txt"), "a") as file:
    file.write(
f"""T: {lable}\n  
P: {np.argmax(preds, axis=-1)}\n\n""")
```

```python
# convert the true and predicted labels into tensors
correct_labels = tf.concat([item for item in y_true], axis = 0)
predicted_labels = tf.concat([item for item in y_pred], axis = 0)
```

```python
report = classification_report(predicted_labels, correct_labels, output_dict=True)
```

```python
df_classification_report = pd.DataFrame(report).transpose()
df_classification_report.to_csv(os.path.join("/content", "drive", "MyDrive", "AI", "fonts", "models_results", "model_3_3", "model_3_3_biger_dataset_classification_report_1.csv"))
```
